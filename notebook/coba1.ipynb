```python
import pandas as pd
df=pd.read_csv(r'D:\projects\analisis\data\sales.csv',parse_dates=['date'])
```

```python
df.info()
df.head()
```

```python
df['revenue']=df['quantity']*df['price']
```

```python
df
```

penglaris

```python
hasil1=df.groupby('product')['revenue'].sum().reset_index().sort_values(by='revenue',ascending=False)
```

```python
hasil1
```

```python
import matplotlib.pyplot as plt
hasil1.plot(kind='bar',x='product',y='revenue')
plt.title('product terlaris')
plt.show()
```
